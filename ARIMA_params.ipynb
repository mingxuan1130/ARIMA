What is ACF:

The autocorrelation function (ACF) is a tool used in statistics to quantify the degree to which a data series is correlated with itself over different time intervals or lags. Simply put, it measures how closely current values in a time series are related to past values.

It can help to identify whether a time series has seasonality or cycles.

ACF Plot: This is a visual representation of autocorrelation across different lags. Each bar represents the autocorrelation at a specific lag. Bars outside the confidence interval (often shaded blue or grey) indicate significant autocorrelation at those lags.


What is PACF:

The Partial Autocorrelation Function (PACF) is a tool that helps understand the relationship between different points in a time series. It looks at how much two points in the series, separated by a certain number of periods (called "lags"), are related to each other. The key thing about PACF is that it takes into account and removes the influence of all the points that are closer together than these two points. This means it shows the unique connection between just these two points, without being affected by the other nearby points in the series.

The PACF is a key tool in time series analysis, used alongside the ACF to identify the appropriate terms in an ARIMA model.

Generally, an MA(q) process will cut off at lag q on the ACF and tail off on the PACF. It's the reverse for an AR(p). An AR(p) will cut off at lag p on the PACF and tail off on the ACF. If both the ACF and PACF are tailing, this suggests an ARMA process.

ACF Plot: We can observe the gradual decay of autocorrelation as the lags increase. This is typical for an AR(1) process, where each value in the series is primarily influenced by its immediate predecessor, and this influence diminishes over time.

PACF Plot: There is a significant spike at lag 1, followed by a sharp cut-off. This indicates that the first lag is significant while the subsequent lags are not, which is characteristic of an AR(1) process. The PACF plot essentially confirms the presence of a single significant autoregressive term in the time series.
